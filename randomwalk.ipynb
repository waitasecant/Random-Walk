```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(p,size):
    walk = []
    for i in range(size):
        if np.random.binomial(1,p,size=1) == 1:
            walk.append(1)
        else:
            walk.append(-1)
    return walk
```

```python
def cumsum(x):
    cumsum = []
    for i in range(1,len(x)+1):
        cumsum.append(sum(x[:i]))
    return cumsum
```

```python
def prob_rw(p,size):
    hist = []
    prob = []
    for i in range(1000):
        hist.append(cumsum(random_walk(p,size))[-1])
    for j in range(-size,size+1,2):
        prob.append(hist.count(j)/len(hist))
    return hist, prob
```

```python
plt.figure(figsize=(15,3))
for i,p in zip(range(1,6),[0.1,0.25,0.5,0.75,0.9]):
    size = 10
    hist, prob = prob_rw(p,size)
    plt.subplot(1,5,i)
    plt.vlines(range(-size,size+1,2),[0 for i in range(len(prob))],prob, colors= "red")
    plt.scatter(range(-size,size+1,2),prob)
plt.show()
```
